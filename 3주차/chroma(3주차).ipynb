```python
import os
from langchain.text_splitter import CharacterTextSplitter
from langchain.vectorstores import Chroma
from langchain.embeddings import SentenceTransformerEmbeddings
from langchain.document_loaders import PyPDFLoader
from langchain.document_loaders import TextLoader
from langchain.document_loaders import DirectoryLoader

from langchain.embeddings import OpenAIEmbeddings
from langchain.vectorstores import Chroma
from langchain.chains import ConversationalRetrievalChain
from langchain.memory import ConversationBufferMemory
from langchain.llms import OpenAI

from langchain.embeddings import LlamaCppEmbeddings, HuggingFaceEmbeddings
```

```python
embeddings = HuggingFaceEmbeddings()
```

```python
#구글드라이브 마운트하기
from google.colab import drive
drive.mount('/content/drive')
```

```python
fn_dir = "/content/drive/My Drive/Colab Notebooks/webtoon_text"
```

```python
loader = DirectoryLoader(fn_dir)
documents = loader.load()

text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
docs = text_splitter.split_documents(documents)

len(documents), len(docs)
```

```python
db = Chroma.from_documents(docs, embedding=embeddings,
                                 persist_directory="webtoon_index_hf")
db.persist()
```

```python
query = "화요일에 나오는것"
docs = db.similarity_search(query)
```
